```python
import pandas as pd 
import seaborn as sns

data = pd.read_csv("nba2.csv")
#show first 5 columns
data.head()

```

```python
#plotinglineplot
sns.lineplot(x=data['Age'],y=data['Weight'])
```

```python
sns.lineplot(x=data['Age'],y=data['Weight'],hue=data["Position"])
sns.scatterplot(x=data['Age'],y=data['Weight'],hue=data["Position"])
```

```python
sns.boxplot(x=data['Age'])
sns.boxplot(x=data['Age'],y=data['Weight'])
```
