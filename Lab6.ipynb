```python
import numpy as np
import matplotlib.pyplot as plt

#create some sample data
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

#Line plot
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#scatter plot
plt.scatter(x,y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Bar Chart
plt.bar(x,y)
plt.title('Bar Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#Histogram
y1 = np.array([10,10,300,20,500,60,700,80,900,50,220,450])
plt.hist(y1)
plt.title('Histogram')
plt.xlabel('Bins')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie Chart
labels = ['A','B','C','D','E']
sizes = [15,30,45,10,5]
plt.pie(sizes,labels=labels)
plt.title('Pie Chart')
plt.show()
```
